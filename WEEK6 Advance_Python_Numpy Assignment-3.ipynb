````python
import numpy as np
````

````python
climate_data_station1=np.genfromtxt("climate_data_station1.csv",delimiter=",",skip_header=1,dtype="int",names=["temperature","rainfall","humidity"])
````

````python
climate_data_station2=np.genfromtxt("climate_data_station2.csv",delimiter=",",dtype="int",skip_header=1,names=["temperature","rainfall","humidity"])
````

````python
climate_data_station3=np.genfromtxt("climate_data_station3.csv",delimiter=",",dtype="int",skip_header=1,names=["temperature","rainfall","humidity"])
````

````python
climate_data_station4=np.genfromtxt("climate_data_station4.csv",delimiter=",",dtype="int",skip_header=1,names=["temperature","rainfall","humidity"])
````

````python
climate_data_station5=np.genfromtxt("climate_data_station5.csv",delimiter=",",dtype="int",skip_header=1,names=["temperature","rainfall","humidity"])
````

````python
climate_data_station6=np.genfromtxt("climate_data_station6.csv",delimiter=",",dtype="int",skip_header=1,names=["temperature","rainfall","humidity"])
````

````python
climate_data_station7=np.genfromtxt("climate_data_station7.csv",delimiter=",",dtype="int",skip_header=1,names=["temperature","rainfall","humidity"])
````

````python
climate_data_station8=np.genfromtxt("climate_data_station8.csv",delimiter=",",dtype="int",skip_header=1,names=["temperature","rainfall","humidity"])
````

````python
climate_data_station9=np.genfromtxt("climate_data_station9.csv",delimiter=",",dtype="int",skip_header=1,names=["temperature","rainfall","humidity"])
````

````python
climate_data_station10=np.genfromtxt("climate_data_station10.csv",delimiter=",",dtype="int",skip_header=1,names=["temperature","rainfall","humidity"])
````

````python
climate=np.concatenate((climate_data_station1,climate_data_station2,climate_data_station3,climate_data_station4,climate_data_station5,climate_data_station6,climate_data_station7,climate_data_station8,climate_data_station9,climate_data_station10))
````

````python
climate_data=climate.reshape(-1,1)
````

````python
yield_of_crop=(climate_data["temperature"]*0.3)+(climate_data["rainfall"]*0.2)+(climate_data["humidity"]*0.5)
````

````python
yield_of_crop
````

````python
temperature=climate["temperature"].reshape(-1,1)
````

````python
rainfall=climate["rainfall"].reshape(-1,1)
````

````python
humidity=climate["humidity"].reshape(-1,1)
````

````python
climates_results=np.concatenate((temperature,rainfall,humidity,yield_of_crop),axis=1)
climates_results
````

````python
np.savetxt("climates_results.csv",climates_results,delimiter=",",fmt=['%i','%i','%i','%s'], header="temperature,rainfall,humidity,yield_of_crop") # numpy method
````
